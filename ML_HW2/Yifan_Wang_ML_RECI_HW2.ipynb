```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gradient_descent(gradient,start,learn_rate,iter=50,tolerance=1e-06):
    vector=start;
    all_vector=[vector]
    for _ in range(iter):
        diff=-learn_rate*gradient(vector)
        if np.all(abs(diff) <= tolerance):
            break
        vector+=diff
        all_vector.append(vector)
    return all_vector
```

```python
def f(v):
    return v**4 - 5 * v**2 - 3 * v

all_vectors = gradient_descent(
    gradient=lambda v: 4 * v**3 - 10 * v - 3, start=0, learn_rate=0.2
)
print(f"number of iterations: {len(all_vectors)}")
```

```python
v = np.linspace(-3, 3, int(1e5))
plt.plot(v, f(v))

plt.plot(all_vectors, f(np.array(all_vectors)), marker='o', color='green')
```
